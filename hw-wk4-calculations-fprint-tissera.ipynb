# Creating a Dataframe

```python
import pandas as pd

data = {
    'City': ['Chicago', 'Los Angeles', 'New York City', 'San Jose, CA'],
    'Murders': [762, 294, 334, 47],
    'Population': [2705000, 3976000, 8538000, 1025000]
}

df = pd.DataFrame(data)

df
```

```python
def murderRate(row):
    return (row['Murders'] / row['Population'])*100000

df['Murder Rate'] = df.apply(murderRate, axis=1)
df
```
